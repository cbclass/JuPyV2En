```python
%matplotlib widget

from ipywidgets import interact, interactive, fixed, interact_manual
import matplotlib.pyplot as plt


def plotten(start,stop,step): 
    plt.clf()
    X = []
    Y = []
    xPunkte=[2,4,6,8]
    yPunkte=[22,24,26,28]
    plt.plot(xPunkte,yPunkte,'xr')
    for x in range(start,stop,step):
        X.append(x)
        Y.append(20 + x) 
    plt.plot(X,Y, 'bo')
    plt.xlabel('x values')
    plt.ylabel('y values')
    plt.ylim(20,29)
    plt.xlim(1,9)

a=interactive(plotten, start=(-2,5), stop=(5,9), step=(1,5))
display(a)
```
